```python
import pandas as pd
```

```python
titles = pd.read_csv("../content/cleaned_movie_db.csv")
```

```python
credits = pd.read_csv("../content/cleaned_credits.csv")
```

```python
titles = pd.DataFrame(titles)
credits = pd.DataFrame(credits)
```

```python
imdb_score = titles[titles['imdb_score']>=8.0]
imdb_score.head()
```

```python
imdb_sort = imdb_score.sort_values(['imdb_score'], ascending=False)
imdb_sort.head(5)
```

```python
g = imdb_score["age_certification"].str.contains('G').sum()
pg = imdb_score["age_certification"].str.contains('PG').sum()
pg13 = imdb_score["age_certification"].str.contains('PG-13').sum()
r = imdb_score["age_certification"].str.contains('R').sum()
print (g, pg, pg13, r)
```

```python
average_runtime = imdb_score["runtime"].mean()
average_runtime
```

```python
genre_list = imdb_score["genres"].tolist()
genres = []
genres.append(genre_list)
print (genres)
```

```python
crime = imdb_score["genres"].str.contains('crime').sum()
drama = imdb_score["genres"].str.contains('drama').sum()
horror = imdb_score["genres"].str.contains('horror').sum()
comedy = imdb_score["genres"].str.contains('comedy').sum()
documentation = imdb_score["genres"].str.contains('documentation').sum()
war = imdb_score["genres"].str.contains('war').sum()
romance = imdb_score["genres"].str.contains('romance').sum()
thriller = imdb_score["genres"].str.contains('thriller').sum()
scifi = imdb_score["genres"].str.contains('scifi').sum()
animation = imdb_score["genres"].str.contains('animation').sum()
action = imdb_score["genres"].str.contains('action').sum()
music = imdb_score["genres"].str.contains('music').sum()
reality = imdb_score["genres"].str.contains('reality').sum()
history = imdb_score["genres"].str.contains('history').sum()
sport = imdb_score["genres"].str.contains('sport').sum()
western = imdb_score["genres"].str.contains('western').sum()
family = imdb_score["genres"].str.contains('family').sum()
```

```python
mcredits = imdb_score.merge(credits, on = "id")
mcredits.head(5)
```

```python
credits_sort = mcredits.sort_values(['imdb_score'], ascending=False)
credits_sort.head(5)
```

```python
new_credits_sort = credits_sort.drop(columns=["index_x", "movie", "release_year", "age_certification", "runtime", "genres", "production_countries", "tmdb_popularity", "tmdb_score",
                                              "index_y"])
new_credits_sort.head()
```

```python
new_credits_sort.nunique()
```

```python
actors = new_credits_sort.sort_values('role')
actors2 = actors[actors['role']=="ACTOR"]
actors_sort = actors2.sort_values(['imdb_score'], ascending=False)
actors_sort
```

```python
print(actors_sort['name'].value_counts())
```

```python
directors = new_credits_sort.sort_values('role')
directors2 = directors[directors['role']=="DIRECTOR"]
directors_sort = directors2.sort_values(['imdb_score'], ascending=False)
directors_sort
```

```python
print(directors_sort['name'].value_counts())
```

```python
# Export DataFrames to csv: Movies with the Top IMDB scores
imdb_score.to_csv('Top_scores.csv')

# Export DataFrames to csb: Top Rated actors
actors2.to_csv('Top_actors.csv')

# Exporting Top Directors
directors2.to_csv('Top_Directors.csv')

```
